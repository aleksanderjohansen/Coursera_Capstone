```python
import pandas as pd
```

```python
# importing the data from a csv file stored at github

df=pd.read_csv("https://raw.githubusercontent.com/aleksanderjohansen/Coursera_Capstone/master/NYC.csv")
df.head()
```

```python
# dropping 'Not assigned' values for the Borough column

df.drop(df.index[df['Borough'] == 'Not assigned' ], inplace = True)
df
```

```python
# grouping the data based on postcode and borough, and joining the values for the Neighbourhood column

df.groupby(['Postcode','Borough'], sort = False).agg(lambda x: ','.join(x))
```

```python
# replacing 'Not assigned' values in the Neighbourhood column with vales from the Borough column for each row

df.loc[df['Neighbourhood'] == ('Not assigned'), 'Neighbourhood'] = df['Borough']
```
